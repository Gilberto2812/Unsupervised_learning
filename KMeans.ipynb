```python
#Importing dependencies
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import random as rm
import matplotlib.pyplot as plt
```

```python
#Importing data
from sklearn.datasets import load_iris
all_data = load_iris()['data']

#Create sample
rm.seed(4)
sample = load_iris()['data'][rm.sample(range(150),40)]
```

```python
#Creating an fitting model
model = KMeans(n_clusters=3)
model.fit(sample)
```

```python
#Predict values for sample
pred = model.predict(sample)

#Predict values for all data
pred_all = model.predict(all_data)
```

```python
#Plotting sample - sepal length (cm) vs petal length (cm)
dct_color = {0:'blue', 1:'green', 2:'red'}
gp = plt.scatter(x = sample[:, 0], y = sample[:, 3], c = [dct_color.get(l, 'black') for l in pred])
```

```python
#Plotting all data - sepal width (cm) vs petal length (cm)

gp_all = plt.scatter(x = all_data[:, 1], y = all_data[:, 2], c = [dct_color.get(l, '#9b59b6') for l in pred_all])
```
